```python
import pandas as pd 
import seaborn as sns 
import matplotlib.pyplot as plt 
```

```python
dataset = pd.read_csv(r"C:\Users\Aamir Shahzad\OneDrive\Desktop\archive\2019.csv")
dataset.head()
```

```python
dataset.shape
```

```python
dataset.isnull().sum()
```

```python
# The dataset has already been cleaned and contains no missing or inconsistent values. 
# We can now proceed with visualizing the data through various graphs to explore patterns and trends.
```

```python

# Displaying the top 10 countries with the highest happiness scores using a horizontal bar plot. 
# This helps identify which nations rank highest in overall happiness based on the dataset.

top10 = dataset.sort_values("Score", ascending=False).head(10)
plt.figure(figsize=(6,3))
sns.barplot( data=top10, x="Score", y="Country or region", hue="Score", palette="viridis")
plt.title("Top 10 Happiest Countries")
plt.xlabel("Happiness Score")
plt.ylabel("Countries")
plt.legend(title='Score', loc='center left', bbox_to_anchor=(-0.5, 0.5))
plt.show()
```

```python
# Visualizing the 10 countries with the lowest happiness scores to identify regions with the least reported well-being.
# This plot helps highlight global disparities in happiness and quality of life.

top10_least = dataset.sort_values("Score", ascending=False).tail(10)
plt.figure(figsize=(6,3))
sns.barplot( data=top10_least, x="Score", y="Country or region", hue="Score", palette="rocket")
plt.title("Top 10 Least Happy Countries")
plt.xlabel("Happiness Score")
plt.ylabel("Countries")
plt.legend(title='Score', loc='center left', bbox_to_anchor=(-0.5, 0.5))
plt.show()
```

```python
# This scatter plot shows the connection between GDP per capita and happiness score. 
# It helpsus see if richer countries are generally happier.

plt.figure(figsize=(6,4))
sns.scatterplot(x="GDP per capita", y="Score", data=dataset, hue="Score")
plt.title("GDP per Capita vs Happiness Score")
plt.xlabel("GDP per Capita")
plt.ylabel("Happiness Score")
plt.legend(title='Score', loc='upper left', labelspacing=0.2)
plt.tight_layout()
plt.show()
```

```python
# This heatmap shows how different numerical features are related to each other. 
# It helps us find which factors have a strong link with happiness score.

numeric_df = dataset.select_dtypes(include='number')

sns.heatmap(numeric_df.corr(), annot=True, cmap='coolwarm')
plt.title("Correlation Between Happiness Features")
plt.tight_layout()
plt.show()
```

```python
# This bar plot shows the top 15 countries with the highest generosity scores. 
# It helps us understand which countries are most generous according to the data.

top15_generosity = dataset.sort_values("Generosity", ascending=False).head(15)
sns.barplot(x="Generosity", y="Country or region", data=top15_generosity, palette="viridis", hue="Generosity")
plt.title("Top 15 Most Generous Countries")
plt.tight_layout()
plt.show()
```

```python
# This scatter plot shows the relationship between happiness score and freedom to make life choices.
# It helps us see if people with more freedom feel happier.

sns.scatterplot(x="Score", y="Freedom to make life choices", data=dataset, hue="Freedom to make life choices")
plt.xlabel("Happiness Score")
plt.ylabel("Freedom to make life choices")
plt.title("Freedom vs Happiness Score")
plt.legend(title="Freedom to..." , labelspacing = 0.2)
plt.show()
```

```python
# This scatter plot shows the relationship between healthy life expectancy and happiness score.
# It helps us understand if living longer, healthier lives leads to greater happiness.

sns.scatterplot(data=dataset, x="Healthy life expectancy", y='Score', hue="Healthy life expectancy")
plt.title("Life Expectancy vs Happiness Score")
plt.ylabel("Happiness Score")
plt.show()
```

```python
# This scatter plot shows the relationship between generosity and perceptions of corruption.
# It helps us analyze whether people in less corrupt countries tend to be more generous.

sns.scatterplot(data=dataset, x='Perceptions of corruption', y='Generosity', hue="Generosity")
plt.title("Generosity vs Corruption Perception")
plt.legend(title = "Generosity", loc= "lower center",labelspacing = 0.2,  bbox_to_anchor=(0.58, 0.001))
plt.show()
```

```python
# This pairplot shows the relationships between happiness score and other key factors.
# It helps us visually explore how different features are connected to each other.

sns.pairplot(dataset[['Score', 'GDP per capita', 'Social support', 
                 'Healthy life expectancy', 'Freedom to make life choices', 
                 'Generosity']])
plt.show()
```
